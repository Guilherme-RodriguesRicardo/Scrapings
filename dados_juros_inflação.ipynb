```python
import pandas as pd
import datetime
from bcb import sgs
import pandas_datareader as pdr
import wbdata

pd.options.display.float_format = '{:,.4f}'.format
```

```python
def rolling_index(indices:dict) -> pd.DataFrame: #Usar quando precisar pegar um indice cumulativo
    
    df = sgs.get(indices)
    df.index = df.index.to_period('M')
    df_rolling = df.rolling(12)
    df_cumm = df_rolling.apply(lambda x: (1 + x/100).prod() - 1).dropna() * 100
    name_indice = df_cumm.columns[0]
    df_cumm = df_cumm/100
    value = df_cumm.iloc[-1,0]
    value = "{:.3%}".format(value)
    columns = {'Indices':[name_indice] ,'values':[value]  }
    data = pd.DataFrame(columns)
    return df_cumm

```
